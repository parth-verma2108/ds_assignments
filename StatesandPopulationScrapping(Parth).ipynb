```python
from bs4 import BeautifulSoup
from selenium import webdriver
import unicodedata
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States")
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
table = soup.find('table', 'wikitable')
```

```python
rows = table.find_all('tr')[2:]
```

```python
state_names = []

for i in rows:
    state_names.append(unicodedata.normalize('NFKD', i.find('th').find('a').text).encode('ascii', 'ignore'))
state_names
```

```python
populations = []

for i in rows:
    populations.append(unicodedata.normalize('NFKD', i.find_all('td')[-4].text).encode('ascii', 'ignore'))
populations
```

```python
data = {
    'State': state_names,
    'Population': populations
}
d = pd.DataFrame(data)
d.to_csv('states_and_populations.csv', index=False)
```
